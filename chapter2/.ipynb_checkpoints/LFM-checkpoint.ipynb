```python
import random
import numpy as np
```

```python
# 简单的训练集
train={
    'user0':{
        'item1':5,
        'item2':2,
        'item3':3,
        'item6':5,
    },
    'user1':{
        'item1':5,
        'item0':2,
        'item4':4,
        'item6':3,
    },
    'user2':{
        'item1':5,
        'item3':3,
        'item4':4,
    },
    'user3':{
        'item2':4,
        'item4':2,
        'item5':1,
        'item9':5,
    },
    'user4':{
        'item1':4,
        'item2':5,
        'item8':1,
        'item9':3,
    },
    'user5':{
        'item3':4,
        'item5':2,
        'item0':3,
        'item6':2,
    },
    'user6':{
        'item7':4,
        'item8':3,
        'item3':1,
        'item5':5,
    },
}
```

```python
# 按照物品流行度采样出了热门的但用户没有行为的物品
# 但这个是随机采样，不是按热门但用户没有行为采样
# item_pool:所有item
# user_items:用户正样本
def RandomSelectNegativeSamples(item_pool,user_items):
    ret=dict()
    for i,rating in user_items.items():
        ret[i]=rating
    n=0
    while n<len(user_items):
        item=random.choice(item_pool)
        if item in ret:
            continue
        ret[item]=0
        n+=1
    return ret
```

```python
# 自己写的
def GetItemPool(train):
    item_pool=set()
    for user,items in train.items():
        for i in items.keys():
            item_pool.add(i)
    return list(item_pool)
```

```python
item_pool=GetItemPool(train)
print(item_pool)
```

```python
user_items=train['user1']
print('user1的正负样本')
ret=RandomSelectNegativeSamples(item_pool,user_items)
print(ret)
```

```python
def InitModel(train,F):
    user_num=len(train)
    item_pool=GetItemPool(train)
    item_num=len(item_pool)
    # 初始化P，Q都为值全是2.5的矩阵
    P=np.ones((user_num,F))*2.5
    Q=np.ones((F,item_num))*2.5
    return P,Q
```

```python
def LFM(train, F, N, alpha, lambdaa):
    #初始化P,Q矩阵
    P,Q=InitModel(train,F)
    item_pool=GetItemPool(train)
    #开始迭代
    for step in range(0, N):
        #从数据集中依次取出user以及该user喜欢的iterms集
        for user, items in train.items():
            #随机抽样，为user抽取与items数量相当的负样本，并将正负样本合并，用于优化计算
            samples = RandomSelectNegativeSamples(item_pool,items)
            #依次获取item和user对该item的兴趣度
            for item, rui in samples.items():
                #根据当前参数计算误差
                eui = rui - P.dot(Q)
                u_num=int(user[4])
                i_num=int(item[4])
                for f in range(0, F):
                    
                    P[u_num][f] += alpha * (eui[u_num][i_num] * Q[f][i_num] - lambdaa * P[u_num][f])
                    Q[f][i_num] += alpha * (eui[u_num][i_num] * P[u_num][f] - lambdaa * Q[f][i_num])
        #每次迭代完后，都要降低学习速率。一开始的时候由于离最优值相差甚远，因此快速下降；
        #当优化到一定程度后，就需要放慢学习速率，慢慢的接近最优值。
        alpha *= 0.9
    return P,Q
```

```python
F=5
alpha=0.02
lambdaa=0.01
N=100

P,Q=LFM(train,F,N,alpha,lambdaa)
```
